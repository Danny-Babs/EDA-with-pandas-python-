```python
#Seasons for the data -> are the races slower in summer than in winter ?

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2
```

```python
df2['race_month'] = df2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df2['race_season'] = df2['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else'Spring' if x > 2 else 'Winter' )
```

```python
df2['avg_speed'] = pd.to_numeric(df2['avg_speed'], errors='coerce')  # Convert avg_speed to numeric
result = df2.groupby('race_season')['avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
